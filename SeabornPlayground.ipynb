```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CSV_data/london_houses.csv', encoding='latin-1')
```

```python
df.head()
```

```python
sns.scatterplot(x='Square Meters', y='Price', data=df, hue='Garden', style='Garage')
```

```python
df['Neighborhood'].unique()
```

```python
sns.countplot(x='Neighborhood', data=df)
plt.xticks(rotation=90)
```

```python
df.info()
```

```python
sns.catplot(x='Interior Style', col='Property Type', kind='count', data=df, hue='Interior Style', legend=None)
```

```python
sns.relplot(x='Building Age', y='Square Meters', data=df, size='Price', hue='Price')
```

```python
g=sns.catplot(x='Bedrooms', y='Price', data=df, kind='point', errorbar=None, hue='Bathrooms', palette='Set2', linestyle='--')
```
